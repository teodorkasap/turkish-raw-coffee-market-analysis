## Modules

```python
import pandas as pd
import numpy as np
```

```python
file1 = '2000-2019-ithalat-ihracat-rakamlari.xls'
table1 = pd.read_excel(file1, skiprows=list(range(1)))
```

# Raw Table

```python
table1
```

```python
table1['year']=table1['year'].astype(str)
table1['export']=table1['export'].replace('\.','',regex=True)
table1['export']=table1['export'].astype(float)*1000
table1['import']=table1['import'].replace('\.','',regex=True)
table1['import']=table1['import'].astype(float)
table1['export value']=table1['export value'].replace('\.','',regex=True)
table1['export value']=table1['export value'].astype(float)*1000
table1['import value']=table1['import value'].replace('\.','',regex=True)
table1['import value']=table1['import value'].astype(float)

table1
```

### Calculate Local Consumption and Merge to Initial Table

```python
table1["local"]=table1["import"]-table1["export"]
table1
```

### Calculate Average Imp. & Exp. Prices

```python
table1["av. exp. value"]=table1["export value"]/table1["export"]
table1["av. imp. value"]=table1["import value"]/table1["import"]
table1
```
